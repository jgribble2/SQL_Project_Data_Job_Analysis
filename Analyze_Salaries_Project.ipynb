```python
# import libraries
import pandas as pd
from datasets import load_dataset

# load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# clean data - convert column to datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# view countries with the most jobs postings
df['job_country'].value_counts().head(20)
```

```python
# check if a random country is in the postings
# use .isin() and .any()
df['job_country'].isin(['South Sudan']).any()
```

```python
# filter the data to just Canadian jobs
canada_jobs = df[df['job_country'] == "Canada"]
canada_jobs
```

```python
# filter out all the NA salaries with the .na() method
canada_jobs = canada_jobs[canada_jobs['salary_year_avg'].notna()]
```

```python
# aggregate the job_title_short column by the count of jobs (salary_year_avg column) with the .groupby() method
canada_jobs.groupby('job_title_short')['salary_year_avg'].count()
```

```python
# use the .agg() method to view the median, max, min, and count of salaries, and then sort by the median with the .sort_values() method
canada_jobs.groupby('job_title_short')[['salary_year_avg']].agg(['median', 'min', 'max', 'count']).sort_values(by = 'median')
# error comes up, that's because when using the .sort_value, you must remove the list that the aggregated column (salary_year_avg) is in
```

```python
# have the results descending by median (add 'ascending = False')
canada_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by = 'median', ascending = False)
```
